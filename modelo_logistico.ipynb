Efecto Allee

```
def modelo_allee(l,k,nt,i):
  NtUno = []
  Nt = []
  print('N(0) = '+str(nt))
  for s in range(i):
    Nt.append(nt)
    nt = nt + nt**2*(k-nt)*(nt-l)
    NtUno.append(nt)
    print('N('+str(s+1)+')'+' = '+str(nt))
  #maimax = max(max(Nt), max(NtUno))
  plt.style.use('dark_background')
  fig= plt.figure(figsize=(13,8))
  plt.scatter(Nt, NtUno)
  plt.plot(Nt, NtUno)
  plt.plot([0, k], [0, k])
  plt.ylabel('N(t+1)')
  plt.xlabel('N(t)')
  plt.grid()
  plt.show()
```

#1. Modelo de Ricker

```
def modelo_ricker(vals, rango_max, r, k):

  rango = [0]

  if vals > 2:
    saltos = rango_max/(vals-1)
    s = saltos
    for i in range(vals-2):
      rango.append(s)
      s+=saltos
    rango.append(rango_max)
  elif vals == 2:
    rango.append(rango_max)

  NtUno = []
  #print('N(0) = '+str(nt))
  for s in range(len(rango)):
    #Nt.append(nt)
    nt = rango[s]*math.exp(r*(1-rango[s]/k))
    NtUno.append(nt)
    #print('N('+str(s+1)+')'+' = '+str(nt))
  #maimax = max(max(Nt), max(NtUno))
  plt.style.use('dark_background')
  fig= plt.figure(figsize=(13,8))
  plt.title('Modelo de Ricker')
  plt.scatter(rango, NtUno)
  plt.plot(rango, NtUno)
  plt.plot([0, k], [0, k])
  plt.ylabel('N(t+1)')
  plt.xlabel('N(t)')
  plt.grid()
  plt.show()

  #return rango
```

```
#Rango máximo = K
modelo_ricker(100,50,1.2,50)
```

```
#Rango_máximo > K
modelo_ricker(100,100,1.2,50)
```

#1. Modelo logístico

```
def modelo_logistico(vals, rango_max, r, k):

  rango = [0]

  if vals > 2:
    saltos = rango_max/(vals-1)
    s = saltos
    for i in range(vals-2):
      rango.append(s)
      s+=saltos
    rango.append(rango_max)
  elif vals == 2:
    rango.append(rango_max)

  NtUno = []
  #print('N(0) = '+str(nt))
  for s in range(len(rango)):
    #Nt.append(nt)
    nt = rango[s] + r*rango[s]*(1-rango[s]/k)
    NtUno.append(nt)
    #print('N('+str(s+1)+')'+' = '+str(nt))
  #maimax = max(max(Nt), max(NtUno))
  plt.style.use('dark_background')
  fig= plt.figure(figsize=(13,8))
  plt.title('Modelo logistico')
  plt.scatter(rango, NtUno)
  plt.plot(rango, NtUno)
  plt.plot([0, k], [0, k])
  plt.ylabel('N(t+1)')
  plt.xlabel('N(t)')
  plt.grid()
  plt.show()
```

```
#Rango_máximo = K
modelo_logistico(100,50,1.2,50)
```

```
#Rango_máximo > K
modelo_logistico(100,100,1.2,50)
```

Cuando el rango máximo es igual a K, ambos presentan una curva que termina tocando la identidad. Siendo más pronunciada la curva generada por el modelo de Ricker

Por otro lado, cuando el rango máximo es mayor a K, se observa que después de la indentidad el modelo de Ricker presenta una recta con pendiente negativa pero que sigue estando en los positivos. En cambio, con el modelo logistico presentan una pendiente negativa pronunciada con una curva que cuando en el modelo de Ricker dio 30, en este regresó -20

#2. Puntos de equilibrio

```
def modelo_equilibrio(r,k,nt,i):
  NtUno = []
  Nt = []
  print('N(0) = '+str(nt))
  for s in range(i):
    Nt.append(nt)
    nt = nt*math.exp(r*(1-nt/k))
    NtUno.append(nt)
    print('N('+str(s+1)+')'+' = '+str(nt))
  #maimax = max(max(Nt), max(NtUno))
  plt.style.use('dark_background')
  fig= plt.figure(figsize=(13,8))
  plt.scatter(Nt, NtUno)
  plt.plot(Nt, NtUno)
  plt.plot([0, k], [0, k])
  plt.ylabel('N(t+1)')
  plt.xlabel('N(t)')
  plt.grid()
  plt.show()
```

```
#Si empezamos con N(0) = 0, no cambia la población. Lógico
modelo_equilibrio(1.2,10,0,10)
```

```
#Con N(0) = 10, la población se mantiene constante
modelo_equilibrio(1.2,10,10,10)
```

```
#Aquí del 5 se va al punto de equilibrio 10
modelo_equilibrio(1.2,10,5,20)
```

```
#Aquí se ve que 0 es un punto de equilirbio inestable, mandando a 10 la población inicial en tan solo 3 iteraciones
modelo_equilibrio(1.2,10,1,10)
```

```
#Número cercano a 0, que también se va después de unas cuantas iteraciones a 10
#Y por ejemplo, aquí se ve que termina quedandose en 10 de manera permanente
modelo_equilibrio(1.2,10,0.1,100)
```
